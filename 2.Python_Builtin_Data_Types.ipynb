# Python Built-in Data Types

## What are the built-in data types in Python?

Python provides several built-in data types categorized into the following groups:

### 1. Numeric types:
- `int` – Integer values
- `float` – Floating point numbers
- `complex` – Complex numbers

### 2. Sequence types:
- `str` – Strings
- `list` – Ordered, mutable sequences
- `tuple` – Ordered, immutable sequences

### 3. Set types:
- `set` – Unordered collections of unique elements
- `frozenset` – Immutable version of a set

### 4. Mapping types:
- `dict` – Key-value pairs

### 5. Boolean type:
- `bool` – Boolean values (`True` or `False`)

### 6. Binary types:
- `bytes`, `bytearray`, `memoryview`

### 7. None type:
- `NoneType` – Represents the absence of a value (e.g., `None`)


```
# Numeric types
a = 10        # int
b = 3.14      # float
c = 2 + 3j    # complex

# Sequence types
s = "Hello"           # str
lst = [1, 2, 3]       # list
tup = (4, 5, 6)       # tuple

# Set types
st = {1, 2, 3}        # set
fst = frozenset([1, 2, 3])  # frozenset

# Mapping type
d = {'name': 'Alice', 'age': 25}  # dict

# Boolean type
flag = True           # bool

# Binary types
bt = b'hello'         # bytes
ba = bytearray(5)     # bytearray
mv = memoryview(bytes(5))  # memoryview

# None type
n = None              # NoneType

# Display all data types
print(type(a), type(b), type(c))
print(type(s), type(lst), type(tup))
print(type(st), type(fst))
print(type(d))
print(type(flag))
print(type(bt), type(ba), type(mv))
print(type(n))
```
